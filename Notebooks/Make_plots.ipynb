```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Data
df = pd.read_csv('Data/mentions.csv', header=0)
df['x_values'] = ['Top 10', 'Top 30', 'Top 50', 'Top 100']
df
```

```python
# multiple line plots
plt.plot( 'x_values', 'Accuracy Mean', data=df, marker='o', markerfacecolor='blue', markersize=5, color='skyblue', linewidth=1)
plt.plot( 'x_values', 'Precision Mean', data=df, marker='o', markerfacecolor='blue', markersize=5, color='olive', linewidth=1)
plt.plot( 'x_values', 'Recall Mean', data=df, marker='o', markerfacecolor='blue', markersize=5, color='green', linewidth=1)
plt.plot( 'x_values', 'F1 Mean', data=df, marker='o', markerfacecolor='blue', markersize=5, color='red', linewidth=1)
# show legend
plt.legend()
plt.rcParams['axes.facecolor'] = 'black'

plt.title('Overview mean of mention feature')
# show graph
plt.savefig('Data/mean_mention.png',bbox_inches='tight')
plt.show()
```

```python
# multiple line plots
plt.plot( 'x_values', 'Acurracy Std', data=df, marker='o', markerfacecolor='blue', markersize=5, color='skyblue', linewidth=1)
plt.plot( 'x_values', 'Precision Std', data=df, marker='o', markerfacecolor='blue', markersize=5, color='olive', linewidth=1)
plt.plot( 'x_values', 'Recall Std', data=df, marker='o', markerfacecolor='blue', markersize=5, color='green', linewidth=1)
plt.plot( 'x_values', 'F1 Std', data=df, marker='o', markerfacecolor='blue', markersize=5, color='red', linewidth=1)
# show legend
plt.legend()
plt.title('Overview standard deviation of all the featuers combined')
# show graph
plt.savefig('Data/std_all.png',bbox_inches='tight')

plt.show()
```
